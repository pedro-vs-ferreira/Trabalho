### __Exercício 2.1__


```python
import pandas as pd

# Dados das probabilidades
data = {
    'p1(H)': [0.5, 0.6, 0.4, 0.1, 0.3],
    'p2(H)': [0.5, 0.7, 0.8, 0.2, 0.4]
}

df = pd.DataFrame(data)

# Calculando as probabilidades dos resultados
df['HH'] = df['p1(H)'] * df['p2(H)']
df['HT'] = df['p1(H)'] * (1 - df['p2(H)'])
df['TH'] = (1 - df['p1(H)']) * df['p2(H)']
df['TT'] = (1 - df['p1(H)']) * (1 - df['p2(H)'])

# Imprimindo a tabela 
print(df)
```

### __Exercicío 2.2__

```python
import pandas as pd

# Probabilidades iniciais
total_emails = 963
prob_spam = 0.6  # Probabilidade de um email ser spam
prob_nao_spam = 0.4  # Probabilidade de um email não ser spam
prob_acerto_spam = 0.98  # Probabilidade do filtro identificar corretamente um spam
prob_acerto_nao_spam = 0.02  # Probabilidade do filtro identificar incorretamente um spam como não spam

# Calculando a probabilidade de um email ser marcado como spam
prob_marcado_como_spam = (prob_acerto_spam * prob_spam) + (prob_acerto_nao_spam * prob_nao_spam)

# Calculando a probabilidade de um email NÃO ser spam, dado que foi marcado como spam
# Usamos o Teorema de Bayes
prob_nao_spam_dado_marcado_como_spam = (prob_acerto_nao_spam * prob_nao_spam) / prob_marcado_como_spam

# Calculando o número esperado de emails não spam que foram marcados como spam
num_nao_spam_marcados_como_spam = prob_nao_spam_dado_marcado_como_spam * total_emails

print("Número estimado de emails não spam na caixa de spam:", num_nao_spam_marcados_como_spam)
```

### __Exercicío 2.3__

```python
def decifrar_cesar(texto, deslocamento):
    """Decifra uma cifra de César."""
    resultado = ""
    for letra in texto:
        if letra.isalpha():
            indice = (ord(letra) - ord('A') - deslocamento) % 26
            resultado += chr(indice + ord('A'))
        else:
            resultado += letra
    return resultado

texto_cifrado = "OCZ XJMMZXO VINRZM"
deslocamento = 7

# Decifra a mensagem 3 vezes
for _ in range(3):
    texto_cifrado = decifrar_cesar(texto_cifrado, deslocamento)

print("Mensagem decifrada:", texto_cifrado)
```

### __Exercicío 2.4__

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados da distribuição
data = {'Resultado': [1, 2, 3, 4, 5, 6, 7, 8],
        'Probabilidade': [0.1, 0.15, 0.2, 0.21, 0.17, 0.09, 0.06, 0.02]}

# Criando um DataFrame
df = pd.DataFrame(data)

# Calculando a CDF
df['CDF'] = df['Probabilidade'].cumsum()

# Visualizando a PMF
plt.bar(df['Resultado'], df['Probabilidade'], label='PMF')
plt.xlabel('Resultado')
plt.ylabel('Probabilidade')
plt.title('Função de Massa de Probabilidade')
plt.show()

# Visualizando a CDF
plt.step(df['Resultado'], df['CDF'], label='CDF')
plt.xlabel('Resultado')
plt.ylabel('Probabilidade Acumulada')
plt.title('Função de Distribuição Acumulada')
plt.show()
```

### __Exercicío 2.5__

```python
import pandas as pd
from scipy.stats import entropy
from sklearn.metrics import mutual_info_score

# Carrega os dados do arquivo 'data.txt'
df = pd.read_csv('data.txt', sep='\t')

# Calcula a entropia (informação) de cada variável
# A entropia mede a incerteza ou a quantidade de informação média de uma variável aleatória
v1_ps = (df.groupby(by='v1').size() / df.shape[0]).values
v2_ps = (df.groupby(by='v2').size() / df.shape[0]).values
v1_info = entropy(v1_ps)
v2_info = entropy(v2_ps)

print(v1_info)
print(v2_info)

# Calcula a informação mútua entre v1 e v2
# A informação mútua mede a quantidade de informação que uma variável fornece sobre a outra
mi = mutual_info_score(df['v1'], df['v2'])

# Calcula a porcentagem de informação de v1 que é explicada por v2
print(f"Conhecer v2 nos da {100 * mi / v1_info:.2f}% de v1")
```

### __Exercicío 4.2__

```python
import networkx as nx
from collections import defaultdict
import matplotlib.pyplot as plt
G = nx.read_edgelist("dat@.txt", create_using=nx.MultiGraph(), data=[("tipo de aresta", str)])
edge_types_to_analyze = ["C"]
node_edge_counts = defaultdict(lambda: defaultdict(int))
for u, v, data in G.edges(data=True):
    edge_type = data["tipo de aresta"]
    if edge_type in edge_types_to_analyze:
        node_edge_counts[u][edge_type] += 1
        node_edge_counts[v][edge_type] += 1
pos = nx.spring_layout(G) 
nx.draw(G, pos, with_labels=True)
plt.show()
```

```python
import networkx as nx

G = nx.read_edgelist("d@dos.txt", create_using = nx.MultiGraph(), data = [("start", int), ("end", int)])

first_window = nx.Graph()
second_window = nx.Graph()
third_window = nx.Graph()
for e in G.edges(data = True):
    if e [2]["start"] <= 3:
        first_window.add_edge(e[0], e[1])
    if e [2]["start"] <= 6 and e[2]["end"] > 3:
        second_window.add_edge(e[0], e[1])
    if e [2]["end"] > 6:
        third_window.add_edge(e[0], e[1])

pos = nx.spring_layout(G) 
nx.draw(G, pos, with_labels=True)
plt.show()
```
